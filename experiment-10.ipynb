```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import myipf # 前面的实验编写的函数
from matplotlib.gridspec import GridSpec
#from mpl_toolkits.mplot3d.axes3d import Axes3D

%matplotlib inline
plt.rcParams["font.sans-serif"] = ["SimHei"] # 用来正常显示中文
plt.rcParams["axes.unicode_minus"]=False # 用来正常显示负号
%config InlineBackend.figure_format = "svg"
```

```python
img = plt.imread("standard_test_images/lena_gray_256.tif")
```

## 图像比特平面

```python
plt.imshow(myipf.bitPlane(img, 1, "bit", True), cmap="gray")
```

### 
我们可以看到LSB置0或置1均方差很小，都不超过1，峰值信噪比51dB，与原图十分相似，人眼无法分辨。  
但是把第4位置0或置1，峰值信噪比只有33dB左右，人眼可察觉到变化，图像甚至出现了伪轮廓。

```python
LSB_set_one = myipf.bitPlane(img, 1, "one")
LSB_set_zero = myipf.bitPlane(img, 1, "zero")
four_set_one = myipf.bitPlane(img, 4, "one")
four_set_zero = myipf.bitPlane(img, 4, "zero")
```

```python
plt.figure
plt.subplot(1, 3, 1), plt.imshow(img, cmap="gray"), plt.axis("off"), plt.title("原图")
plt.subplot(1, 3, 2), plt.imshow(LSB_set_one, cmap="gray"), plt.axis("off"), plt.title("LSB置1")
plt.subplot(1, 3, 3), plt.imshow(LSB_set_zero, cmap="gray"), plt.axis("off"), plt.title("LSB置0")
plt.show()
myipf.calcMSEPSNR(img, LSB_set_one)
myipf.calcMSEPSNR(img, LSB_set_zero)
```

```python
plt.figure
plt.subplot(1, 3, 1), plt.imshow(img, cmap="gray"), plt.axis("off"), plt.title("原图")
plt.subplot(1, 3, 2), plt.imshow(four_set_one, cmap="gray"), plt.axis("off"), plt.title("第4位置1")
plt.subplot(1, 3, 3), plt.imshow(four_set_zero, cmap="gray"), plt.axis("off"), plt.title("第4位置0")
plt.show()
myipf.calcMSEPSNR(img, four_set_one)
myipf.calcMSEPSNR(img, four_set_zero)
```
